```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
print(df.columns)
```

```python
plt.figure(figsize=(10, 6))
plt.bar(['male', 'female'], [df['male'].sum(), df['female'].sum()], color=['blue', 'pink'])
plt.xlabel('Gender')
plt.ylabel('Population')
plt.title('Male vs Female Population in India')
plt.show()
```

```python
states = df['state/ut']
male_population = df['male']
female_population = df['female']

plt.figure(figsize=(12, 10))
plt.bar(states, male_population, color='skyblue', label='Male')
plt.bar(states, female_population, color='pink', label='Female', bottom=male_population)
plt.xlabel('State/UT')
plt.ylabel('Population')
plt.title('Gender Distribution by State/UT in India')
plt.xticks(rotation=90)
plt.legend()
plt.tight_layout()
plt.show()
```
