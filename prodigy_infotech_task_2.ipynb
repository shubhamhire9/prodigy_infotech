```python
new_data=dataset.drop(columns=['Trades','Deliverable Volume','%Deliverble','Series'])
new_data
```

```python
correlation_matrix = new_data.corr(numeric_only=True)
print(correlation_matrix)
dataplot = sns.heatmap(correlation_matrix, cmap="YlGnBu", annot=True)
plt.show()
```

```python
sns.pairplot(new_data)
plt.show()
```

```python
x=dataset['Date'].tail(25)
y=dataset['Prev Close'].tail(25)
fig=plt.figure(figsize=(18,6))
plt.subplot(1,2,2)
plt.plot(x,y,marker='v',color='red')
plt.tick_params(axis='x',rotation=90)
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Last 25 Days Trend')
#fig.set_figwidth(10)
x=dataset['Date'].head(25)
y=dataset['Prev Close'].head(25)
fig.set_figwidth(10)
plt.subplot(1,2,1)
plt.plot(x,y,marker='^',color='green')
plt.tick_params(axis='x',rotation=90)
plt.title('First 25 Days Trend')
plt.xlabel('Date')
plt.ylabel('Price')
fig=plt.figure()
#fig.set_figwidth(10)
plt.tight_layout()
plt.show()
```

```python
volume=new_data['Volume'].head(500)
sns.scatterplot(volume)
plt.title('Scatter plot of volume traded between range')
plt.xlabel('Range of Days')
plt.ylabel('Total Number of Shares Traded')
plt.show()
```

```python
volume=new_data['Volume'].head(500)
sns.displot(volume, kde = True, color = "g")
plt.tick_params(axis='x',rotation=90)
plt.title('Distribution of volume')
plt.show()
```

```python
volume=new_data['Close'].head(500)
sns.displot(volume, kde = True, color = "r")
plt.tick_params(axis='x',rotation=90)
plt.title('Distribution of Close Price')
plt.show()
```

```python
print(np.linspace(161,2570,num=5,endpoint=True))
r161_763=dataset[(dataset['VWAP']>=161)&(dataset['VWAP']<=763)]
r764_1365=dataset[(dataset['VWAP']>=764)&(dataset['VWAP']<=1365)]
r1366_1967=dataset[(dataset['VWAP']>=1366)&(dataset['VWAP']<=1967)]
r1968_2570= dataset[(dataset['VWAP']>=1968)&(dataset['VWAP']<=2570)]

range=[len(r161_763),len(r764_1365),len(r1366_1967),len(r1968_2570)]
labels=['161 to 763','764 to 1365','1366 to 1967','1968 to 2570']

sns.barplot(x=labels,y=range,hue=labels)
plt.title('Box plot of VWAP range')
plt.xlabel('Ranges of VWAP')
plt.ylabel('VWAP Value')
plt.show()
```
