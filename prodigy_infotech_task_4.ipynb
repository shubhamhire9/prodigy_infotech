```python
import pandas as pd
from textblob import TextBlob
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
twitter_data = pd.read_csv("/content/twitter_training.csv")
```

```python
twitter_data['sentiment'] = twitter_data['text'].apply(lambda x: TextBlob(str(x)).sentiment.polarity)
```

```python
plt.hist(twitter_data['sentiment'], bins=30, color='green', edgecolor='black')
plt.title('Sentiment Distribution of twitter data')
plt.xlabel('Sentiment Polarity')
plt.ylabel('Frequency')
plt.show()
#data is encoded thats why we cannot perform more visualization operation
```

```python
# restructuring the positive and negative text data with space (' ').
positive_text = ' '.join(twitter_data[twitter_data['sentiment'] > 0]['text'])
negative_text = ' '.join(twitter_data[twitter_data['sentiment'] < 0]['text'])
```

```python
#visualizing the analyzed data with plt.imshow()
plt.figure(figsize=(18, 10))
plt.subplot(1, 2, 1)
positive_words = WordCloud(width=1000, height=600, background_color='white').generate(positive_text)
plt.imshow(positive_words, interpolation='bilinear')
plt.title('Positive Sentiment Words')
plt.axis('off')
plt.subplot(1, 2, 2)
negative_words = WordCloud(width=1000, height=600, background_color='white').generate(negative_text)
plt.imshow(negative_words, interpolation='bilinear')
plt.title('Negative Sentiment Words')
plt.axis('off')
plt.show()
```
